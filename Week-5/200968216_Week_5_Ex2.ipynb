### Q1. Perform the preprocessing if required, scale the train and test data using standard scaler.


```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
```

```python
df = pd.read_csv('HousePrice.csv')
```

```python
df.tail(5)
```

```python
df.describe()
```

```python
df.info()
```

Removing Alley, FireplaceQu ,PoolQC ,Fence and MiscFeature as they have too many NULL values

```python
df = df.drop(['Alley', 'FireplaceQu', 'PoolQC', 'Fence', 'MiscFeature'], axis=1)
```

Removing ID as its irrelevant

```python
df = df.drop(['Id'], axis=1)
```

```python
df.info()
```

```python
df.describe()
```

#### Removing rows having more than 3 null values from dataset

```python
df = df.dropna(thresh=3)
```

```python
df.describe()
```

Making list of categorical and numeric variables for pre-processing

```python
numerical_cols   = [x for x in df.columns if df[x].dtype in ['int64', 'float64']]
categorical_cols = [x for x in df.columns if df[x].dtype == 'object']
```

Removing SalePrice as its the dependent variable.

```python
numerical_cols.remove('SalePrice')
```

Imputing missing values using SimpleImputer replacing numerical null to mean while categorial as most frequent.

Also doing onehot encoding and Standard Scaling Dataset.

```python
num_imputer = SimpleImputer(strategy='mean')
cat_imputer = SimpleImputer(strategy='most_frequent')

numerical_transformer = Pipeline(steps = [('scaler', StandardScaler()),('imputer', num_imputer)])
categorical_transformer = Pipeline(steps=[('onehot', OneHotEncoder(handle_unknown='ignore')),('imputer', cat_imputer)])
preprocessor = ColumnTransformer(transformers=[('num', numerical_transformer, numerical_cols),('cat', categorical_transformer, categorical_cols),])
```

```python
y = df.SalePrice
X = df.drop(['SalePrice'], axis=1)
X.head()
```

### Q2. Split the dataset into train size of 70% and test size of 30% and Apply the Ridge and Lasso regression and fit the model containing all independent variables.


```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
lasso = Lasso(alpha=0.1)
pipe_l = Pipeline(steps=[('preprocessor', preprocessor), ('model', lasso)])
```

```python
pipe_l.fit(X_train, y_train)
```

```python
pipe_l.score(X_test, y_test)
```

```python
ridge = Ridge(alpha=1.0)

pipe_r = Pipeline(steps=[('preprocessor', preprocessor), 
                              ('model', ridge)])
```

```python
pipe_r.fit(X_train, y_train)
```

```python
pipe_r.score(X_test, y_test)
```
